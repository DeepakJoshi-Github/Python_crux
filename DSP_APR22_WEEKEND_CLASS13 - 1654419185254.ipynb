```python
def func( x ):
    plt.figure( figsize = (5, 3) )
    plt.title( 'This is a boxplot for ' + x.name )
    x.plot(kind = 'box')
    plt.show()

stores.select_dtypes('float').dropna( axis = 1, how = 'all' ).apply( func )
plt.show()
```

```python
plt.figure( figsize = (5, 3) )
sns.boxplot( y = stores.TotalSales)
plt.show()
```

```python
summ = stores.StoreType.value_counts()
plt.bar( x = summ.index, height = summ.values )
plt.show()
```

```python
sns.countplot( y = stores.StoreType )
```

#### Bi variate analysis

continuous, continuous: scatter plots / xyplots

categorical, categorical: crosstab, heatmap

continuous, categorical: bar, barh, line, area, pie, doughnut

```python
stores.plot( x = 'TotalSales', y = 'OperatingCost', kind = 'scatter')
plt.show()
```

```python
stores.plot( x = 'TotalSales', y = 'Total_Customers', kind = 'scatter')
plt.show()
```

```python
sns.scatterplot( x = stores.TotalSales, y = stores.TotalSales )
```

```python
sns.pairplot( stores.select_dtypes(['float', 'int64']).dropna(axis = 1, how = 'all') )
plt.show()
```

```python
stores.select_dtypes(['float', 'int64']).dropna(axis = 1, how = 'all').corr().to_excel('D:/Corr.xlsx')
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap( stores.select_dtypes(['float', 'int64']).dropna(axis = 1, how = 'all').corr(), annot = True )
plt.savefig('d:/heatmap.jpeg')
```

```python
plt.scatter( x = stores.TotalSales, y = stores.OperatingCost, c = '#80ff80', s = 50, marker = '+' )
plt.show()

# marker: ^, +, *, .
```

```python
sns.scatterplot( x = stores.TotalSales, y = stores.OperatingCost, hue = stores.Location, s = 100 )
```

```python
pd.crosstab( index = stores.StoreType, columns = stores.StoreSegment )
```

```python
sns.heatmap( pd.crosstab( stores.StoreType, stores.StoreSegment ), annot=True )
```

```python
stores.groupby('Location').TotalSales.sum().plot( kind = 'bar' )
```

```python
stores.groupby('Location').TotalSales.sum().plot( kind = 'pie' )
```

#### multi variate analysis

Clustered Charts / Stacked Charts / 100% stacked charts

```python
stores.pivot_table( index = 'Location', columns = 'StoreType', 
                   values = 'TotalSales', aggfunc = 'sum').plot( kind = 'bar' )
```

```python
stores.pivot_table( index = 'Location', columns = 'StoreType', 
                   values = 'TotalSales', aggfunc = 'sum').plot( kind = 'bar', stacked = True )
```

```python
summ = stores.pivot_table( index = 'Location', columns = 'StoreType', 
                   values = 'TotalSales', aggfunc = 'sum')
```

```python
summ
```

```python
summ.div( summ.sum( axis = 1 ), axis = 0 ) 
```

```python
summ.div( summ.sum( axis = 1 ), axis = 0).plot( kind = 'bar', stacked = True )
```

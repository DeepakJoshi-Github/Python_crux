### summarize the data

```python
group by

.groupby()
.sum(), .mean(), .min(), .... .agg()

To create a cross table from agg vertical table
.pivot()

pivot tables, sumif/s, .....
.pivot_table()

To create only the cross tables
pd.crosstab()
```

```python
import numpy as np
import pandas as pd
```

```python
stores = pd.read_csv('D:/SampleData/stores_new.csv')
```

```python
stores.head()
```

```python
# Get the the Location wise total sales
summ1 = stores.groupby( by = 'Location' ).TotalSales.sum()
```

```python
summ1
```

```python
summ1.name
```

```python
# Get the the Location wise total sales and total operating cost
stores.loc[:, ['Location', 'TotalSales', 'OperatingCost']].groupby('Location').sum()
```

```python
stores.groupby('Location')[['TotalSales', 'OperatingCost']].sum().reset_index()
```

```python
grp_by = stores.groupby('Location')
```

```python
grp_by.TotalSales.sum()
```

```python
grp_by[['TotalSales', 'OperatingCost']].sum()
```

```python
# get the Total and Average of TotalSales by Location
stores.groupby('Location').TotalSales.agg(['sum', 'mean', 'min', 'max', 'count'])
```

```python
stores.groupby('Location').TotalSales.agg([np.sum, np.mean, min, max])
```

```python
# Get the Total and Average of TotalSales and OperatingCost
stores.groupby('Location')[['TotalSales', 'OperatingCost']].agg([np.sum, np.mean, min, max])
```

```python
# Get the Total TotalSales and Average of OperatingCost
stores.groupby('Location')[['TotalSales', 'OperatingCost']].agg( {'TotalSales': 'sum', 
                                                                  'OperatingCost': 'mean' } )
```

```python
stores.groupby('Location')[['TotalSales', 
        'OperatingCost']].agg( {'TotalSales': ['sum', 'mean', 'min', 'max'], 
                                                'OperatingCost': ['sum', 'mean'] } )
```

```python
# get the Location and StoreType wise TotalSales
summ = stores.groupby(['Location', 'StoreType']).TotalSales.sum().reset_index()
summ
```

```python
summ.pivot( index = 'Location', columns = 'StoreType', values = 'TotalSales' )
```

```python
stores.pivot_table( index = 'Location', columns = 'StoreType', values = 'TotalSales', aggfunc = 'sum' )
```

```python
stores.pivot_table( index = ['Location', 'StoreType'], values = 'TotalSales', aggfunc = ['sum', 'mean'] )
```

```python
stores.columns
```

```python
# get the count of records for StoreType and StoreSegment
stores.pivot_table( index = 'StoreType', columns= 'StoreSegment', 
                           values = 'StoreCode', aggfunc = 'count' )
```

```python
pd.crosstab( index = stores.StoreType, columns = stores.StoreSegment )
```

### Combining the datasets

```python
file_path = 'D:/SampleData/'
```

```python
demog = pd.read_csv( file_path + 'Demographic_Data.csv')
trans = pd.read_csv( file_path + 'Transaction_Summary.csv')
```

```python
demog
```

```python
trans
```

```python
pd.merge( left = demog, right = trans, how = 'inner', left_on = 'CustName', right_on = 'CustomerName' )
```

```python
pd.merge( left = demog, right = trans, how = 'left', left_on = 'CustName', right_on = 'CustomerName' )
```

```python
pd.merge( left = demog, right = trans, how = 'right', left_on = 'CustName', right_on = 'CustomerName' )
```

```python
pd.merge( left = demog, right = trans, how = 'outer', left_on = 'CustName', right_on = 'CustomerName' )
```

```python
patient = pd.read_excel( file_path + 'Demog.xlsx' )
visit = pd.read_excel( file_path + 'Visits.xlsx' )
```
